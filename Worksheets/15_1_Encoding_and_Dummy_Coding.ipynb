## Encoding Data  
When we encode data, we remodel a numeric column in a dataframe to be on a standard scale (0 or 1).   
For example if we had a column of BMI scores, we could encode that column so that all scores greater than or equal to 25 were recoded to the value 1 (bad) and all scores less than 25 were recoded to 0 (good).  

For example:  

` def encode_bmi(df):`       
> `if df['bmi'] >= 25:`  
> >  `return 1`   

>`else`:  
> >`return 0`  

`df["bmi"] = df.apply(encode_bmi, axis=1)`

````python
#=======================================================================
# encode daily vaccinations per million, where values greater than or 
# equal to median = 1 and values less than median = 0
#=======================================================================
def encode_daily(df, **kwds):
  median = kwds['median']
  key = kwds['key']

  if df[key] >= median:
    return 1
  else:
    return 0

#=============================================================================
# Start
#=======================================================================
# get the median daily_vaccinations_per_million
#=======================================================================
median = get_median(covid_vax_df)
covid_vax_df["encoded"] = covid_vax_df.apply(encode_daily, axis = 1, key="daily_vaccinations_per_million", median=median)

covid_vax_df['encoded'].describe()

````

````python
#=======================================================================
# Task 2 - encode total_vaccinations col:
# all values < UK min are 0 and otherwise coded as 1
#=======================================================================
def encode_total(df, **kwds):
  total = kwds['minimum']
  key = kwds['key']

  if df[key] >= total:
    return 1
  else:
    return 0

#=======================================================================
# Task 3 - Display the unique countries for which total vaccinated is 
# at the same rate or more than the UK
#=======================================================================
def display_unique_countries(df):
   
  unique_country = df.groupby('total_vaccinations_encoded').apply(lambda x: x['country'].unique())

  print(unique_country[1])

  return 

#=============================================================================
# Start
#=======================================================================
# Task 1 - find the min UK total_vaccinations 
#=======================================================================

minimum = covid_vax_df[covid_vax_df['country'] == 'United Kingdom'].total_vaccinations.min()

print("\nMin UK total Vax: ", minimum)

#=======================================================================
# Task 2 - encode total_vaccinations col:
# all values < UK min are 0 and otherwise coded as 1
#=======================================================================
covid_vax_df["total_vaccinations_encoded"] = covid_vax_df.apply(encode_total, 
                            axis = 1, key="total_vaccinations", minimum=minimum)

display(covid_vax_df['total_vaccinations_encoded'].describe())

#=======================================================================
# Task 3 - Display the unique countries for which total vaccinated is 
# at the same rate or more than the UK
#=======================================================================
display_unique_countries(covid_vax_df )
````

For example:

* to duplicate an existing column

* to add two columns together 
  * `df['new_column'] = df['column1'] + df['column2']`
* to make a percentages column

````python

#=======================================================================
# Task 2 - find the sum of total vaccinations for each manufacturer
#=======================================================================
def calc_sum_per_mnfctr(df):
  sum_total_vax = df[["vaccine", "total_vaccinations"]].groupby("vaccine").sum()
  #print(sum_total_vax.to_string())

  return sum_total_vax

#=======================================================================
# Task 3 - create new col with total vaccinations as a % overall sum of 
# total vaccinations
#=======================================================================
def create_new_col(df):

  df['percent_total_sum_vax'] = (df['total_vaccinations']/df['total_vaccinations'].sum()) * 100
 
  return df
#=======================================================================
# Task 4
# find the median percent_total_sum_vax
#=======================================================================
def calc_median(df):

  median_percent = df['percent_total_sum_vax'].median()
  
  #print("\median_percent = ", median_percent)

  return median_percent

#=======================================================================
# Task 5 - encode percent_total_sum_vax col:
# all values >= median percentage = 1 and otherwise coded as 0
#=======================================================================
def encode_total(df, **kwds):
  total = kwds['median']

  key = kwds['key']

  if df[key] >= total:
    return 1
  else:
    return 0

#=======================================================================
# program starts here
#=======================================================================
# Task 1 - read excel file into df
#=======================================================================

url = "https://github.com/lilaceri/Working-with-data-/blob/342abab10d93c4bf23b5c55a50f189f12a137c5f/Data%20Sets%20for%20code%20divisio/Covid%20Vaccination%20Data.xlsx?raw=true"
covid_vax_by_mnfctr_df = pd.read_excel(url,sheet_name="by_manufacturer")

display (covid_vax_by_mnfctr_df.columns)

#=======================================================================
# Task 2 - find the sum of total vaccinations for each manufacturer
#=======================================================================
sum_total_vax = calc_sum_per_mnfctr(covid_vax_by_mnfctr_df)

print(sum_total_vax.to_string())

#=======================================================================
# Task 3 - create new col with total vaccinations as a % overall sum of 
# total vaccinations
#=======================================================================

covid_vax_by_mnfctr_new_df = create_new_col(covid_vax_by_mnfctr_df)

print("\ncovid_vax_by_mnfctr_new_df: ", covid_vax_by_mnfctr_new_df.columns)

#=======================================================================
# Task 4 - find median percentage of new col percent_total_sum_vax
#=======================================================================

median_percent = calc_median(covid_vax_by_mnfctr_new_df)

print("\nmedian percent: ", median_percent, "\n\n")

#=======================================================================
# Task 5 - encode percent_total_sum_vax col:
#=======================================================================
covid_vax_by_mnfctr_new_df["encoded_percentages"] = covid_vax_by_mnfctr_new_df.apply(encode_total, axis = 1, 
                              key="percent_total_sum_vax", median=median_percent)

display(covid_vax_by_mnfctr_new_df)
````
